### Clustering based fault detection

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate example time series data
np.random.seed(0)
num_samples = 200
time_series = np.sin(np.linspace(0, 4*np.pi, num_samples)) + np.random.normal(0, 0.3, num_samples)

# Reshape data for clustering (required format for scikit-leran)
X = time_series.reshape(-1, 1)

# Number of clusters (can be determined using elbow method or domain knowledge)
num_clusters = 3

# Initialise and fit K-Means clustering model
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(X)

# Assign each data point to a cluster
cluster_labels = kmeans.predict(X)

# Plot the time series data and cluster assignments
plt.figure(figsize=(10,6))
plt.scatter(range(num_samples), X, c=cluster_labels, cmap='viridis')
plt.title("K-Means Clustering for Fault Detection")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```
